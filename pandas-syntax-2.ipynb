```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_names = pd.DataFrame(df_postlong, columns = ['tags', 'title', 'body'])
```

```python
df_names
```

```python
df_names.dtypes
```

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]

for char in spec_chars:
    df_names['tags'] = df_names['tags'].str.replace(char, ' ')
    df_names['title'] = df_names['title'].str.replace(char, ' ')
```

```python
df_names
```

```python
import re as re

def remove_tags(string):
    result = re.sub('<.*?>','',string)
    return result

df_names['body']=df_names['body'].apply(lambda cw : remove_tags(cw))
```

```python
df_names
```

```python
spec_charsb = ["\n","!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]

for char in spec_charsb :
    df_names['body'] = df_names['body'].str.replace(char, ' ')
```
